### Краткое описание проекта
* Quiz Freeze - игра, сборник вопросов в различных форматах и темах, на которые игрокам нужно давать ответы.

2. Старт обучения -> tutorial_start /  опциональный этап

5. Выбор пакетов вопросов -> pack_choice или  training_choice / выбор бесплатного набора вопросов
6. Покупка платных пакетов вопросов ->  факт совершения оплаты за платные вопросы

### 2. Этап: преобразование данных и выборка пользователей по заданным параметрам

```python
# пробуем настроить фильтрацию по дате регистрации - прописываем условия
condition_1 = (events_df['event_type'] == 'registration') # выбрать только регистрацию
condition_2 = (events_df['start_time'] >= '2018-01-01') # начиная от даты
condition_3 = (events_df['start_time'] < '2019-01-01') # заканчивая датой 31-12-2018
cond = condition_1 & condition_2 & condition_3 # объединение условий для фильтровки 
```

```python
# множество user_id, которые были зарегистрированы в интересующий период
user_id_group = set(events_df[cond]['user_id'])
```

```python
# фильтровка только интересующих нас пользователей
events_df = events_df[events_df['user_id'].isin(user_id_group)]
purchase_df = purchase_df[purchase_df['user_id'].isin(user_id_group)]
```

```python
# преобразовываем даты в нужный формат
events_df['start_time'] = pd.to_datetime(events_df['start_time'])
purchase_df['event_datetime'] = pd.to_datetime(purchase_df['event_datetime'])
```

```python
# добавим в purchase_df тип события (пригодится при объединении)
purchase_df['event_type'] = 'purchase'
```

```python
# переименуем столбцы, чтобы не путать с датами событий и оплат
events_df = events_df.rename(columns={'id':'event_id'})
events_df = events_df.rename(columns={'start_time':'event_time'})
purchase_df = purchase_df.rename(columns={'id':'purchase_id'})
purchase_df = purchase_df.rename(columns={'event_datetime':'payment_time'})
```

```python
# oбъединение датафреймов
total_events_df = pd.concat([events_df, purchase_df], sort=False)
total_events_df.info()
```

### 3. Этап: разделение выборки на сегменты - группы пользователей по уровню сложности 

```python
# содержание значений по уровню сложности
total_events_df['selected_level'].unique()
```

```python
# создание множества без пропущенных значений 
level_choice = set(x for x in set(total_events_df['selected_level']) if pd.isnull(x) == False)
display(level_choice)
```

```python
# есть ли пользователи, которые совершили выбор уровня сложности более 1 раза
user = total_events_df[(total_events_df['selected_level'].isin(level_choice))]
user = user.groupby(['user_id'])
user['selected_level'].value_counts().mean()
# среднее значение равно 1, т.е. пользователи, которые выбрали сложность более 1 раза отсутствуют
```

```python
# делаем разделение пользователей исходя из выбранного уровня сложности ('easy', 'medium', 'hard')
easy_users_id = set (total_events_df[total_events_df['selected_level'] == 'easy']['user_id']) 
medium_users_id = set (total_events_df[total_events_df['selected_level'] == 'medium']['user_id']) 
hard_users_id = set (total_events_df[total_events_df['selected_level'] == 'hard']['user_id']) 
```

```python
# проверяем все ли пользователи, которые выбрали уровень сложности, включены в выборку
len (easy_users_id)+len(medium_users_id)+len(hard_users_id) == total_events_df[total_events_df['selected_level'].isin(level_choice)]['user_id'].nunique()
# вроде, всё верно (8342 == 8342)
```

```python
# интересно посмотреть соотношение сколько всего зарегистрированных пользователей доходят до выбора
# % выбора уровня сложности 
per_who_selected_level= total_events_df[total_events_df['selected_level'].isin(level_choice)]['user_id'].nunique()/total_events_df['user_id'].nunique()
number_who_selected_level = total_events_df[total_events_df['selected_level'].isin(level_choice)]['user_id'].nunique()
easy_level = len(easy_users_id)/number_who_selected_level
medium_level=len(medium_users_id)/number_who_selected_level
hard_level= len(hard_users_id)/ number_who_selected_level
print ('Процент пользователей, которые дошли до выбора уровня сложности: {:.2%} ({} чел.)'.format(per_who_selected_level,number_who_selected_level))
print ()
print ('Разделение предпочтений по уровню сложности:')
print ('   ---> easy_level: {:.2%} ({} чел.)'. format(easy_level, len(easy_users_id)))
print ('   ---> medium_level: {:.2%} ({} чел.) <--- наиболее популярный уровень сложности'. format(medium_level, len(medium_users_id)))
print ('   ---> hard_level: {:.2%} ({} чел.)'. format(hard_level, len(hard_users_id)))
```

### 4. Этап: расчет % платящих пользователей в зависимости от выбранного уровня сложности

```python
# смотрим сколько пользователей перешли к оплате:
purchase_df_easy = purchase_df[purchase_df['user_id'].isin(easy_users_id)]
purchase_df_medium = purchase_df[purchase_df['user_id'].isin(medium_users_id)]
purchase_df_hard = purchase_df[purchase_df['user_id'].isin(hard_users_id)]

# расчет процентa платящих пользователей в зависимости от выбранного уровня сложности:
easy_payment = purchase_df_easy['user_id'].nunique()/len(easy_users_id)
medium_payment = purchase_df_medium['user_id'].nunique()/len(medium_users_id)
hard_payment = purchase_df_hard['user_id'].nunique()/len(hard_users_id)
```

```python
# средний чек по группам
easy_mean=round(purchase_df_easy['amount'].mean(),2)
medium_mean=round(purchase_df_medium['amount'].mean(),2)
hard_mean=round(purchase_df_hard['amount'].mean(),2)
```

```python
print ('Платящие пользователи после выбранного уровня сложности:')
print ()
print ('--> easy_level {:.2%} /{} чел., средний чек: {} руб.'. format(easy_payment,purchase_df_easy['user_id'].nunique(), easy_mean))
print ('-->  medium_level {:.2%} / {} чел., средний чек: {} руб. <---'. format(medium_payment, purchase_df_medium['user_id'].nunique(), medium_mean))
print ('-->  hard_level {:.2%} / {} чел., средний чек: {} руб. <-- наибольший процент оплат после выбора уровня сложности'. format(hard_payment, purchase_df_hard['user_id'].nunique(), hard_mean))
```

### 5. Этап: расчет разницы во времени между регистрацией и оплатой в зависимости от сегментов (п.3)

```python
# напишем условие для фильтрации user_id по по уровню сложности и дате регистрации пользователя
easy_registration_df = total_events_df[(total_events_df['event_type'] == 'registration') & (total_events_df['user_id'].isin(easy_users_id))][['user_id','event_time']]
easy_registration_df = easy_registration_df.rename(columns={'event_time':'registration_time'})

medium_registration_df = total_events_df[(total_events_df['event_type'] == 'registration') & (total_events_df['user_id'].isin(medium_users_id))][['user_id','event_time']]
medium_registration_df = medium_registration_df.rename(columns={'event_time':'registration_time'})

hard_registration_df = total_events_df[(total_events_df['event_type'] == 'registration') & (total_events_df['user_id'].isin(hard_users_id))][['user_id','event_time']]
hard_registration_df = hard_registration_df.rename(columns={'event_time':'registration_time'})
```

```python
# напишем условие для фильтрации user_id по уровню сложности и дате оплаты
easy_purchase_df = purchase_df[(purchase_df['user_id'].isin(easy_users_id))][['user_id', 'payment_time']]
medium_purchase_df = purchase_df[(purchase_df['user_id'].isin(medium_users_id))][['user_id', 'payment_time']]
hard_purchase_df = purchase_df[(purchase_df['user_id'].isin(hard_users_id))][['user_id', 'payment_time']]
```

```python
# объединим датафреймы по user_id
easy_reg_and_purchase_df = easy_registration_df.merge(easy_purchase_df, on =['user_id'], how ='inner')
medium_reg_and_purchase_df = medium_registration_df.merge(medium_purchase_df, on =['user_id'], how ='inner')
hard_reg_and_purchase_df = hard_registration_df.merge(hard_purchase_df, on =['user_id'], how ='inner')
```

```python
# рассчитаем разницу во времени между регистрацией и оплатой для каждой группы пользователей
easy_reg_and_purchase_df['timedelta_reg_to_payment'] = easy_reg_and_purchase_df['payment_time'] - easy_reg_and_purchase_df ['registration_time']
easy_reg_and_purchase_df['level'] = 'easy' #дополним столбец с уровнем сложности для последующего объединения


medium_reg_and_purchase_df['timedelta_reg_to_payment'] = medium_reg_and_purchase_df['payment_time'] - medium_reg_and_purchase_df ['registration_time']
medium_reg_and_purchase_df['level'] = 'medium'

hard_reg_and_purchase_df['timedelta_reg_to_payment'] = hard_reg_and_purchase_df['payment_time'] - hard_reg_and_purchase_df ['registration_time']
hard_reg_and_purchase_df['level'] = 'hard'

#  объединим дата фреймы для удобности отображения мин, макс, среднего и медианы 
total_by_level_df= pd.concat([easy_reg_and_purchase_df,medium_reg_and_purchase_df,hard_reg_and_purchase_df], ignore_index=True)

# выведем статистические показатели для каждой группы (для удобства перевернем столбцы)
total_by_level_df.groupby(by= ['level'])['timedelta_reg_to_payment'].describe().T 
```

### 6. Этап: расчет разницы во времени между выбором уровня и оплатой в зависимости от сегментов

```python
# напишем условие для фильтрации user_id по уровню сложности и дате выбора уровня сложности

easy_select_df = total_events_df[(total_events_df['selected_level'] == 'easy') & (total_events_df['user_id'].isin(easy_users_id))][['user_id','event_time']]
medium_select_df = total_events_df[(total_events_df['selected_level'] == 'medium') & (total_events_df['user_id'].isin(medium_users_id))][['user_id','event_time']]
hard_select_df = total_events_df[(total_events_df['selected_level'] == 'hard') & (total_events_df['user_id'].isin(hard_users_id))][['user_id','event_time']]
```

```python
# объединим датафреймы по user_id (дата выбора уровня + дата оплаты)

easy_select_and_purchase_df = easy_select_df.merge(easy_purchase_df, on ='user_id', how = 'inner')
medium_select_and_purchase_df = medium_select_df.merge(medium_purchase_df, on ='user_id', how = 'inner')
hard_select_and_purchase_df = hard_select_df.merge(hard_purchase_df, on ='user_id', how = 'inner')
```

```python
# рассчитаем разницу во времени для каждой группы пользователей
easy_select_and_purchase_df['timedelta_select_payment'] = easy_select_and_purchase_df['payment_time'] - easy_select_and_purchase_df['event_time']
easy_select_and_purchase_df['level'] = 'easy'

medium_select_and_purchase_df['timedelta_select_payment'] = medium_select_and_purchase_df['payment_time'] - medium_select_and_purchase_df['event_time']
medium_select_and_purchase_df['level'] = 'medium'


hard_select_and_purchase_df['timedelta_select_payment'] = hard_select_and_purchase_df['payment_time'] - hard_select_and_purchase_df['event_time']
hard_select_and_purchase_df['level'] = 'hard'

# объединим датафреймы и сгруппируем для более удобного отображения показателей

total_differ_between_level_and_payment_df = pd.concat([easy_select_and_purchase_df, medium_select_and_purchase_df, hard_select_and_purchase_df], ignore_index = True)
total_differ_between_level_and_payment_df.groupby(by= ['level'])['timedelta_select_payment'].describe().T 
```
