```python
def get_median(mass):
    sort = sorted(mass)
    print(sort)
    if len(sort)%2 == 0:
        med = (sort[len(sort)//2] + sort[len(sort)//2 - 1])/2
        print(len(sort)//2)
    else:
        med = sort[len(sort)//2]
        print(len(sort)//2)
    print(med)
    return med
```

```python
ma = [1,2,2,3,5,7,11,14,14,121,19,20,23,25]
get_median(ma)
```

```python
from random import randint

get_euro_rate = randint(65, 85)
def to_euro(price):  
    exchange_rate = get_euro_rate
    print(exchange_rate)
    rounded = round(price/exchange_rate, 2)  
    return '€' + str(rounded)
```

```python
to_euro(450)
```

```python
def count_letters(string, average=False):
    i = 0
    k = 1
    for letter in string:
        if letter != " ":
            i += 1
        else:
            k += 1
    if average == False:
        return i
    else:
        return i/k
```

```python
count_letters("Beep boop") 
```

```python
count_letters("Beep boop", average=True)
```

```python
count_letters("I will build my own theme park")
```

```python
count_letters("I will build my own theme park", average=True) 
```
