```python
import pandas as pd

events_df = pd.read_csv('events.csv')
eventsf = events_df.loc[(events_df["start_time"] >= "2018-01-01T00:00:00") & (events_df["start_time"] <= "2018-31-12T00:00:00") & (events_df["event_type"] == 'registration')]
registered = eventsf.loc[eventsf["event_type"] == 'registration']["user_id"]
events = events_df[events_df.user_id.isin(registered)]
events.info()
```

```python
registered_count = events[events['event_type'] == 'registration']['user_id'].nunique()
```

```python
tutorial_start_count = events[events['event_type'] == 'tutorial_start']['user_id'].nunique()
```

```python
percent_start = tutorial_start_count/registered_count
print ('Процент пользователей, начавших обучение (от общего числа зарегистрировавшихся): {:.2%}'.format(percent_start))
```

```python
tutorial_finish_count = events[events['event_type'] == 'tutorial_finish']['user_id'].nunique()
percent_finish = tutorial_finish_count/tutorial_start_count
print ('Процент пользователей, закончивших обучение (от общего числа зарегистрировавшихся): {:.2%}'.format(percent_finish))
```

```python
level_choice_count = events[events['event_type'] == 'level_choice']['user_id'].nunique()
percent_level_choice = level_choice_count/registered_count
print ('Процент пользователей, выбравших уровень сложности тренировок (от общего числа зарегистрировавшихся): {:.2%}'.format(percent_level_choice))
```

```python
pack_choice_count = events[events['event_type'] == 'pack_choice']['user_id'].nunique()
percent_pack_choice = pack_choice_count/level_choice_count
print ('Процент пользователей, выбравших набор бесплатных вопросов (от числа пользователей, которые выбрали уровень сложности): {:.2%}'.format(percent_pack_choice))
```

```python
total_events_df = total_events_df.reset_index(drop=True).sort_values('start_time')
total_events_df.start_time = pd.to_datetime(total_events_df.start_time, format='%Y-%m-%dT%H:%M:%S')
```

```python
user_path_df = total_events_df.groupby(['user_id'])['event_type'].apply(list).reset_index()
user_path_df.head(10)
```

```python
user_path_df['event_path'] = user_path_df['event_type'].apply(lambda x: ' > '.join(x))
user_path_df['event_path'].head()
```

```python
user_paths = user_path_df.groupby(['event_path'])['user_id'].nunique().sort_values(ascending=False)
```

```python
registration_df = total_events_df[total_events_df['event_type'] == 'registration']
```

```python
registration_df['user_id'].value_counts().mean()
```

```python
registration_df = registration_df[['user_id','start_time']].rename(columns={'start_time':'registration_time'})
```

```python
registration_df.head()
```

```python
tutorial_start_df = total_events_df[total_events_df['event_type'] == 'tutorial_start']
```

```python
tutorial_start_df['user_id'].value_counts().mean()
```

```python
tutorial_start_df_wo_duplicates = tutorial_start_df.sort_values('start_time').drop_duplicates('user_id')
```

```python
tutorial_start_df_wo_duplicates = tutorial_start_df_wo_duplicates[['user_id','tutorial_id','start_time']].rename(columns={'start_time':'tutorial_start_time'})
```

```python
merged_df = registration_df.merge(tutorial_start_df_wo_duplicates,on='user_id',how='inner')
merged_df.registration_time = pd.to_datetime(merged_df.registration_time, format='%Y-%m-%dT%H:%M:%S')
merged_df.tutorial_start_time = pd.to_datetime(merged_df.tutorial_start_time, format='%Y-%m-%dT%H:%M:%S')
merged_df.head()
```

```python
merged_df['timedelta'] = merged_df['tutorial_start_time'] - merged_df['registration_time']
merged_df.head()
```

```python
merged_df['timedelta'].mean()
```

```python
merged_df['timedelta'].describe()
```

```python
tutorial_finish_df = total_events_df[total_events_df['event_type'] == 'tutorial_finish']
tutorial_finish_df.head()
```

```python
first_tutorial_ids = tutorial_start_df_wo_duplicates['tutorial_id'].unique()
```

```python
tutorial_finish_df = tutorial_finish_df[tutorial_finish_df['tutorial_id'].isin(first_tutorial_ids)]
```

```python
tutorial_finish_df['user_id'].value_counts().mean()
tutorial_finish_df.head()
```

```python
tutorial_finish_df = tutorial_finish_df[['user_id','tutorial_finish_time']]
merged_df_2 = tutorial_start_df_wo_duplicates.merge(tutorial_finish_df,on='user_id',how='inner')
merged_df_2.tutorial_finish_time = pd.to_datetime(merged_df_2.tutorial_finish_time, format='%Y-%m-%dT%H:%M:%S')
merged_df_2.tutorial_start_time = pd.to_datetime(merged_df_2.tutorial_start_time, format='%Y-%m-%dT%H:%M:%S')
merged_df_2['timedelta'] = merged_df_2['tutorial_finish_time'] - merged_df_2['tutorial_start_time']
merged_df_2.head()
print(merged_df_2['timedelta'].mean())
print(merged_df_2['timedelta'].describe())
```

```python
level_choice_df = total_events_df[total_events_df['event_type'] == 'level_choice']
print(level_choice_df['user_id'].value_counts().mean())
level_choice_df = level_choice_df[['user_id','start_time']].rename(columns={'start_time':'level_choice_time'})
merged_df_3 = registration_df.merge(level_choice_df,on='user_id',how='inner')
merged_df_3['timedelta'] = merged_df_3['level_choice_time'] - merged_df_3['registration_time']
print(merged_df_3['timedelta'].mean())
print(merged_df_3['timedelta'].describe())
```

```python
pack_choice_df = total_events_df[total_events_df['event_type'] == 'pack_choice']
print(pack_choice_df['user_id'].value_counts().mean())
pack_choice_df = pack_choice_df[['user_id','start_time']].rename(columns={'start_time':'pack_choice_time'})
merged_df_4 = merged_df_3.merge(pack_choice_df,on='user_id',how='inner')
merged_df_4['timedelta'] = merged_df_4['pack_choice_time'] - merged_df_4['level_choice_time']
print(merged_df_3['timedelta'].mean())
print(merged_df_3['timedelta'].describe())
merged_df_4.head()
```

```python
purchase_df = total_events_df[total_events_df['event_type'] == 'purchase']
print(purchase_df['user_id'].value_counts().mean())
purchase_df.head()
purchase_df = purchase_df[['user_id','event_datetime','amount']].rename(columns={'event_datetime':'purchase_time'})
purchase_df.purchase_time = pd.to_datetime(purchase_df.purchase_time, format='%Y-%m-%dT%H:%M:%S')
merged_df_5 = merged_df_4.merge(purchase_df,on='user_id',how='inner')
merged_df_5['timedelta'] = merged_df_5['purchase_time'] - merged_df_4['pack_choice_time']
print(merged_df_5['timedelta'].mean())
print(merged_df_5['timedelta'].describe())
merged_df_5.head()
```
